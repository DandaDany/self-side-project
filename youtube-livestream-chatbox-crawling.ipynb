# 只抓大俠版本

```python
大俠金句 = {'日期':[],'時間': [], '使用者': [], '訊息': []}
```

```python
for i in range(3600):
    try:
        # 獲取當前日期和時間
        current_time = datetime.now().time()
        
        current_datetime = datetime.now()

        # 提取日期和時間部分
        current_date = current_datetime.date()

        # 提取年、月、日
        year = current_date.year
        month = current_date.month
        day = current_date.day

        # 找到 iframe
        iframe = driver.find_element(By.CSS_SELECTOR, 'iframe[id*="chatframe"]')

        # 切換到 iframe
        driver.switch_to.frame(iframe)

        # 使用 WebDriverWait 等待新留言出現
        wait = WebDriverWait(driver, 10)
        wait.until(EC.presence_of_element_located((By.XPATH, '//span[@id="message"]')))

        # 取得 iframe 內容
        iframe_content = driver.page_source

        # 切換回主要的 frame
        driver.switch_to.default_content()

        # 使用 BeautifulSoup 解析 iframe 內容
        soup = BeautifulSoup(iframe_content, 'html.parser')

        # 提取所有帶有 id=message 或 id=author-name 的元素
        message_elements = soup.find_all('span', {'id': 'message'})
        author_elements = soup.find_all('span', {'id': 'author-name'})
        time_elements = soup.find_all('span', {'id': 'timestamp'})

        # 檢查是否找到相應的元素
        if message_elements and author_elements and len(message_elements) == len(author_elements):
            for message_element, author_element, time_element in zip(message_elements, author_elements, time_elements):
                author_text = author_element.text
                if author_text == '大俠武林':
                    message_text = message_element.text
                    time_text = time_element.text

                    # 檢查是否為新留言
                    if message_text not in 大俠金句['訊息']:
                        大俠金句['日期'].append(f'{year}/{month}/{day}')
                        大俠金句['時間'].append(time_text)
                        大俠金句['使用者'].append(author_text)
                        大俠金句['訊息'].append(message_text)
   
                    else:
                        print('大俠不在')

    except Exception as e:
        print(f"Error: {e}")

    #移動滑鼠
    pyautogui.moveRel(1, 0)
    
    # 等待一段時間，可以根據實際情況調整
    time.sleep(10)
        
    # 檢查是否在凌晨三點之前
    if current_time > datetime.strptime("18:27:00", "%H:%M:%S").time():
        break
```
